1.Data Preparation and Exploratory Analysis

Objective: Load raw transit data,clean and resample into a proper time series,engineer contextual features, and decompose the series to understand its structure before modeling

Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
import warnings
warnings.filterwarnings('ignore')
sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (14, 5)
print('Libraries loaded!')
```

Load Data

```python
df = pd.read_csv('ttc-bus-delay-data-2022.csv')
print(f'Shape : {df.shape}')
print(f'Columns : {df.columns.tolist()}')
print(f'\nData Types:')
print(df.dtypes)
df.head(10)
```

3.Data Cleaning

```python
print('Missing values before cleaning:')
print(df.isnull().sum())
print(f'\nDuplicate rows: {df.duplicated().sum()}')
df = df.dropna(subset=['Min Delay', 'Date', 'Time'])
df = df[df['Min Delay'] >= 0]
df = df[df['Min Delay'] <= 180]
df = df.drop_duplicates()
print(f'\nShape after cleaning: {df.shape}')
print(f'\nMin Delay statistics:')
print(df['Min Delay'].describe().round(2))
```

4.Build Datetime Index

Combining Date and Time columns into a single datetime column

```python
df['datetime'] = pd.to_datetime(
    df['Date'].astype(str) + ' ' + df['Time'].astype(str),
    errors='coerce'
)
failed = df['datetime'].isnull().sum()
print(f'Rows with failed datetime parsing: {failed}')
df = df.dropna(subset=['datetime'])
df = df.sort_values('datetime').reset_index(drop=True)
print(f'\nDate range: {df["datetime"].min()}  →  {df["datetime"].max()}')
print(f'Total records: {len(df):,}')
df[['datetime', 'Route', 'Incident', 'Min Delay', 'Direction', 'Day']].head(10)
```

5. Resample to Hourly Time Series

The dataset has one row per delay incident.We aggregate into a regular hourly time series by taking the mean delay per hour

```python
df_ts = df.set_index('datetime')[['Min Delay']]
ts = df_ts['Min Delay'].resample('1H').mean()
print(f'Total hourly bins  : {len(ts)}')
print(f'Missing bins       : {ts.isnull().sum()}')
print(f'Missing percentage : {ts.isnull().mean()*100:.1f}%')
ts = ts.fillna(0)
print(f'After fill — missing: {ts.isnull().sum()}')
fig, ax = plt.subplots(figsize=(14, 4))
ts.plot(ax=ax,
        color='#457B9D',
        linewidth=0.7,
        alpha=0.9)
ax.set_title('Toronto TTC Bus Delay — Hourly Mean (2022)',
             fontsize=14,
             fontweight='bold')
ax.set_ylabel('Avg Delay (minutes)')
ax.set_xlabel('Date')
plt.tight_layout()

import os
os.makedirs('reports', exist_ok=True)

plt.savefig('reports/fig1_raw_series.png',
            dpi=150,
            bbox_inches='tight')
plt.show()
print(f'\nTime series stats:')
print(ts.describe().round(2))
```

6.Stationarity Test-Augmented Dickey-Fuller

ARIMA requires the series to be statinary.The ADF test checks this formally.
    p<0.05->stationary
    p>=0.05->non_stationary,differencing needed

```python
import os
from statsmodels.tsa.stattools import adfuller
os.makedirs('reports', exist_ok=True)
plt.savefig('reports/fig1_raw_series.png',
            dpi=150,
            bbox_inches='tight')
def adf_test(series, name='Series'):
    result = adfuller(series.dropna(), autolag='AIC')
    print(f'=== ADF Test: {name} ===')
    print(f'ADF Statistic : {result[0]:.4f}')
    print(f'p-value       : {result[1]:.6f}')
    print(f'Lags used     : {result[2]}')
    print('Critical Values:')
    for k, v in result[4].items():
        print(f'   {k}: {v:.4f}')
    if result[1] < 0.05:
        print('✅ STATIONARY — p < 0.05, reject null hypothesis')
    else:
        print('❌ NON-STATIONARY — p >= 0.05, differencing required')
    print()
    return result[1]
p_val = adf_test(ts, 'Hourly Mean Delay')
if p_val >= 0.05:
    print('Applying first-order differencing...')
    ts_diff = ts.diff().dropna()
    adf_test(ts_diff, 'Differenced Series')
else:
    ts_diff = ts
    print('No differencing needed — series is already stationary.')
```

7.Time Series Decomposition

Decomposes the series into Trend,Seasonality, and Residual components.

```python
ts_sample = ts[:24 * 60]
decomp = seasonal_decompose(ts_sample, model='additive', period=24)
fig, axes = plt.subplots(4, 1, figsize=(14, 11))
pairs = [
    (ts_sample,        'Observed',              '#1D3557'),
    (decomp.trend,     'Trend',                 '#457B9D'),
    (decomp.seasonal,  'Seasonality (Daily)',   '#E63946'),
    (decomp.resid,     'Residual (Noise)',      '#2A9D8F'),
]
for ax, (data, label, color) in zip(axes, pairs):
    data.plot(ax=ax, color=color, linewidth=0.9)
    ax.set_ylabel(label, fontweight='bold', fontsize=11)
    ax.set_xlabel('')
axes[0].set_title('Time Series Decomposition — Toronto TTC Bus Delays 2022',
                  fontsize=14, fontweight='bold')
import os
os.makedirs('reports', exist_ok=True)
plt.tight_layout()
plt.savefig('reports/fig2_decomposition.png',
            dpi=150,
            bbox_inches='tight')
plt.show()
```

OBSERVATION:
Trend:Reveals gradual changes in average delay over the year
Seasonality:Clear 24-hour pattern-delays peak at AM(7-9) and PM(16-19) rush hours every day
Residual:Random noise left after removing trend and seasonality-ideally white noise

9.Feature Engineering

```python
df_feat = ts.to_frame(name='delay_minutes').copy()
df_feat['hour']        = df_feat.index.hour
df_feat['dayofweek']   = df_feat.index.dayofweek
df_feat['month']       = df_feat.index.month
df_feat['is_weekend']  = (df_feat['dayofweek'] >= 5).astype(int)
df_feat['is_rush_am']  = ((df_feat['hour'] >= 7) & (df_feat['hour'] <= 9)).astype(int)
df_feat['is_rush_pm']  = ((df_feat['hour'] >= 16) & (df_feat['hour'] <= 19)).astype(int)
df_feat['is_monday']   = (df_feat['dayofweek'] == 0).astype(int)
df_feat['is_friday']   = (df_feat['dayofweek'] == 4).astype(int)
df_feat['lag_1']   = df_feat['delay_minutes'].shift(1)    # 1 hour ago
df_feat['lag_2']   = df_feat['delay_minutes'].shift(2)    # 2 hours ago
df_feat['lag_3']   = df_feat['delay_minutes'].shift(3)    # 3 hours ago
df_feat['lag_24']  = df_feat['delay_minutes'].shift(24)   # same hour yesterday
df_feat['lag_168'] = df_feat['delay_minutes'].shift(168)  # same hour last week
df_feat['rolling_mean_3']  = df_feat['delay_minutes'].rolling(3).mean()
df_feat['rolling_mean_6']  = df_feat['delay_minutes'].rolling(6).mean()
df_feat['rolling_mean_24'] = df_feat['delay_minutes'].rolling(24).mean()
df_feat['rolling_std_6']   = df_feat['delay_minutes'].rolling(6).std()
df_feat = df_feat.dropna()
print(f'Feature set shape : {df_feat.shape}')
print(f'Features created  : {len(df_feat.columns)}')
print(f'\nAll features:')
for col in df_feat.columns:
    print(f'  {col}')
df_feat.head()
```

Statistical Models-ARIMA,SARIMA & Prophet

OBJECTIVE:Train ARIMA,SARIMA,and Prophet on the hourly TTC delay series.Evaluate each model using RMSE and MAE.Plot Actual vs Predicted

1.Setup & Load Data

```python
import joblib
warnings.filterwarnings('ignore')
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.metrics import mean_squared_error, mean_absolute_error
sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (14, 5)
os.makedirs('models', exist_ok=True)
os.makedirs('reports', exist_ok=True)
ts = pd.read_csv('data/processed_timeseries.csv', index_col=0, parse_dates=True)
ts = ts.squeeze()
ts.name = 'delay_minutes'
print(f'✅ Time series loaded: {len(ts):,} hourly rows')
print(f'Range : {ts.index.min().date()}  →  {ts.index.max().date()}')
print(f'Mean  : {ts.mean():.2f} min  |  Max: {ts.max():.2f} min  |  Std: {ts.std():.2f} min')
```

2.Train/Test Split

```python
TEST_HOURS     = 24 * 30   # 30 days holdout
FORECAST_HOURS = 24        # 1 day for visualisation

train = ts[:-TEST_HOURS]
test  = ts[-TEST_HOURS:]

print(f'Train : {len(train):,} hrs  ({train.index[0].date()} → {train.index[-1].date()})')
print(f'Test  : {len(test):,} hrs   ({test.index[0].date()} → {test.index[-1].date()})')

fig, ax = plt.subplots(figsize=(14, 4))
train[-24 * 60:].plot(ax=ax, label='Train (last 60 days shown)', color='#457B9D', linewidth=0.9)
test.plot(ax=ax, label='Test (30 days)', color='#E63946', linewidth=1.3)
ax.set_title('Train / Test Split — TTC Hourly Delays', fontsize=13, fontweight='bold')
ax.set_ylabel('Avg Delay (minutes)')
ax.legend()
import os
os.makedirs('reports', exist_ok=True)
plt.savefig('reports/fig_train_test_split.png',
            dpi=150,
            bbox_inches='tight')
plt.show()
```

3.ACF & PACF PLOTS

These plots help identify the (p,q) parameters for ARIMA:
*ACF:shows how the series correlates with its past values->helps identify q
*PACF:shows direct correlation with past lags->helps identify p

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
plot_acf(train.dropna(),  lags=48, ax=axes[0], color='#457B9D',
         title='ACF — TTC Hourly Delay')
plot_pacf(train.dropna(), lags=48, ax=axes[1], color='#E63946',
          title='PACF — TTC Hourly Delay')
axes[0].set_xlabel('Lag (hours)')
axes[1].set_xlabel('Lag (hours)')
import os
os.makedirs('reports', exist_ok=True)
plt.savefig('reports/fig4_acf_pacf.png',
            dpi=150,
            bbox_inches='tight')
plt.show()
```

4.Auto-ARIMA-Automatically Find Best(p,d,q)

```python
try:
    from pmdarima import auto_arima
    print('Running auto_arima on TTC data (3–5 minutes)...')
    auto_model = auto_arima(
        train,
        start_p=0, max_p=3,
        start_q=0, max_q=3,
        d=None,
        seasonal=False,
        stepwise=True,
        information_criterion='aic',
        suppress_warnings=True,
        error_action='ignore'
    )
    best_order = auto_model.order
    print(f'\n✅ Best ARIMA order: {best_order}')
    print(auto_model.summary())
except ImportError:
    print('pmdarima not installed. Using default order (2, 1, 2).')
    print('Install with: pip install pmdarima')
    best_order = (2, 1, 2)
    print(f'Using order: {best_order}')
```

5.Model 1-ARIMA (Baeline)

```python
print(f'Fitting ARIMA{best_order} on TTC data...')
arima_model = ARIMA(train, order=best_order)
arima_fit   = arima_model.fit()
print(arima_fit.summary())
arima_forecast = arima_fit.forecast(steps=TEST_HOURS)
arima_forecast = np.clip(arima_forecast.values, 0, None)  
arima_rmse = np.sqrt(mean_squared_error(test.values, arima_forecast))
arima_mae  = mean_absolute_error(test.values, arima_forecast)
print(f'\n=== ARIMA{best_order} Results ===')
print(f'RMSE : {arima_rmse:.4f} minutes')
print(f'MAE  : {arima_mae:.4f} minutes')
```

6.MODEL 2-SARIMA

Adds a seasonal component with period=24 to capture TTC's daily rush-hour cycle

```python
print(type(sarima_forecast))
print(locals().keys())
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
SEASONAL_PERIOD = 24
best_order = (1,1,1)
print(f'Fitting SARIMA{best_order}x(1,1,1,{SEASONAL_PERIOD}) on TTC data...')
print('(This may take 5–10 minutes. Run on Colab for speed.)')
train = ts[:-TEST_HOURS]
test  = ts[-TEST_HOURS:]
sarima_model = SARIMAX(
    train,
    order=best_order,
    seasonal_order=(1, 1, 1, SEASONAL_PERIOD),
    enforce_stationarity=False,
    enforce_invertibility=False
)
sarima_fit = sarima_model.fit(disp=False)
sarima_forecast = sarima_fit.forecast(steps=TEST_HOURS)
sarima_forecast = np.clip(sarima_forecast.values, 0, None)
sarima_rmse = np.sqrt(mean_squared_error(test.values, sarima_forecast))
sarima_mae  = mean_absolute_error(test.values, sarima_forecast)
print(f'\n=== SARIMA{best_order}x(1,1,1,{SEASONAL_PERIOD}) Results ===')
print(f'RMSE : {sarima_rmse:.4f} minutes')
print(f'MAE  : {sarima_mae:.4f} minutes')
import pandas as pd
results_df = pd.DataFrame({
    "Actual": list(test),
    "Predicted": list(sarima_forecast)
})
results_df.to_csv("sarima_results.csv", index=False)
print("SARIMA results saved successfully!")
```

7.MODEL 3-FACEBOOK PROPHET

Prophet handles daily+weekly+yearly seasonality automatically.Ideal for TTC which has strong seasonal patterns

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_error
df = pd.read_csv('data/processed_timeseries.csv')
df['datetime'] = pd.to_datetime(df['datetime'])
df_ts = df.set_index('datetime')[['Min Delay']]
ts = df_ts['Min Delay'].resample('1h').mean()
ts = ts.fillna(0)
print('Time series created successfully')
print(ts.head())
split_index = int(len(ts) * 0.8)
train = ts.iloc[:split_index]
test = ts.iloc[split_index:]
TEST_HOURS = len(test)
print('\nTrain size:', len(train))
print('Test size :', len(test))
prophet_train = train.reset_index()
prophet_train.columns = ['ds', 'y']
prophet_model = Prophet(
    daily_seasonality=True,
    weekly_seasonality=True,
    yearly_seasonality=True,
    changepoint_prior_scale=0.05,
    seasonality_mode='additive'
)
prophet_model.fit(prophet_train)
future = prophet_model.make_future_dataframe(
    periods=TEST_HOURS,
    freq='h'
)
prophet_pred = prophet_model.predict(future)
prophet_forecast = prophet_pred['yhat'].values[-TEST_HOURS:]
prophet_forecast = np.clip(prophet_forecast, 0, None)
prophet_rmse = np.sqrt(
    mean_squared_error(test.values,
                       prophet_forecast)
)
prophet_mae = mean_absolute_error(
    test.values,
    prophet_forecast
)
print('\n=== Prophet Results ===')
print(f'RMSE : {prophet_rmse:.4f} minutes')
print(f'MAE  : {prophet_mae:.4f} minutes')
os.makedirs('reports', exist_ok=True)
fig = prophet_model.plot_components(prophet_pred)
plt.suptitle(
    'Prophet Components — TTC Bus Delay',
    fontsize=13,
    fontweight='bold'
)
plt.tight_layout()
plt.savefig(
    'reports/fig5_prophet_components.png',
    dpi=150,
    bbox_inches='tight'
)
plt.show()
```

DEEP LEARNING-LSTM MODEL

Building an LSTM neuralnetwork to capture complex temporal patterns in TTC dlay data.Compare performance against ARIMA,SARIMA and Prophet

```python
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
sns.set_theme(style='whitegrid')
os.makedirs('models', exist_ok=True)
os.makedirs('reports', exist_ok=True)
ts = pd.read_csv('data/processed_timeseries.csv', index_col=0, parse_dates=True)
ts = ts.squeeze()
ts.name = 'delay_minutes'
print(f'✅ TTC time series loaded: {len(ts):,} hourly records')
print(f'Range : {ts.index.min().date()}  →  {ts.index.max().date()}')
print(f'Mean  : {ts.mean():.2f} min  |  Std: {ts.std():.2f} min  |  Max: {ts.max():.2f} min')
```

3.TRAIN/TEST SPLIT

```python
EST_HOURS = 24 * 30   
LOOK_BACK  = 24        
values     = ts.values.reshape(-1, 1).astype(float)
train_vals = values[:-TEST_HOURS]
test_vals  = values[-TEST_HOURS:]
print(f'Train      : {len(train_vals):,} hours')
print(f'Test       : {len(test_vals):,} hours')
print(f'Look-back  : {LOOK_BACK} hours (1 day window)')
```

4.NORMALIZE-MINMAXSCALER

```python
scaler = MinMaxScaler(feature_range=(0, 1))
train_scaled = scaler.fit_transform(train_vals)
test_scaled  = scaler.transform(test_vals)
print(f'Train scaled — min: {train_scaled.min():.3f}  max: {train_scaled.max():.3f}')
print(f'Test  scaled — min: {test_scaled.min():.3f}  max: {test_scaled.max():.3f}')
import os
import joblib
os.makedirs('models', exist_ok=True)
joblib.dump(scaler, 'models/lstm_scaler.pkl')
print('✅ Scaler saved')
print('\n✅ Scaler saved → ../models/lstm_scaler.pkl')
```

5.CREATE A SLIDING WINDOW SEQUENCES

```python
def create_sequences(data, look_back=24):
    """
    Convert a 1D time series into supervised learning format for LSTM.
    Returns X shape: (samples, look_back, 1), y shape: (samples,)
    """
    X, y = [], []
    for i in range(look_back, len(data)):
        X.append(data[i - look_back:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)
all_scaled = np.concatenate([train_scaled, test_scaled], axis=0)
X_all, y_all = create_sequences(all_scaled, LOOK_BACK)
split_idx = len(train_scaled) - LOOK_BACK
X_train, y_train = X_all[:split_idx], y_all[:split_idx]
X_test,  y_test  = X_all[split_idx:], y_all[split_idx:]
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test  = X_test.reshape((X_test.shape[0],  X_test.shape[1],  1))
print(f'X_train : {X_train.shape}  →  (samples, 24 hrs look-back, 1 feature)')
print(f'y_train : {y_train.shape}')
print(f'X_test  : {X_test.shape}')
print(f'y_test  : {y_test.shape}')
```

6.BUILD LSTM MODEL

ARCHITECTURE:2 stacked LSTM layers+Dropout+Dense output layer

```python
if USE_TF:
    tf.random.set_seed(42)
    np.random.seed(42)

    model = Sequential([
        Input(shape=(LOOK_BACK, 1)),

        # Layer 1: LSTM with return_sequences=True for stacking
        LSTM(128, return_sequences=True, activation='tanh',
             recurrent_activation='sigmoid'),
        Dropout(0.2),

        # Layer 2: LSTM — final sequence processing
        LSTM(64, return_sequences=False, activation='tanh',
             recurrent_activation='sigmoid'),
        Dropout(0.2),

        # Dense layers
        Dense(32, activation='relu'),
        Dense(16, activation='relu'),
        Dense(1)   # Output: predicted delay (scaled)
    ])

    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='mse',
        metrics=['mae']
    )

    model.summary()
    print(f'\nTotal parameters: {model.count_params():,}')
```

7.Train the LSTM

```python
import os
os.makedirs('models', exist_ok=True)
if USE_TF:
    callbacks = [
        EarlyStopping(
            monitor='val_loss',
            patience=15,
            restore_best_weights=True,
            verbose=1
        ),
        ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=7,
            min_lr=1e-6,
            verbose=1
        ),
        ModelCheckpoint(
            '..models/lstm_best.h5',
            monitor='val_loss',
            save_best_only=True,
            verbose=0
        )
    ]

    print('🚀 Training LSTM on TTC delay data...')
    history = model.fit(
        X_train, y_train,
        epochs=100,
        batch_size=32,
        validation_split=0.1,
        callbacks=callbacks,
        shuffle=False,   # Important: don't shuffle time series data
        verbose=1
    )

    best_epoch = np.argmin(history.history['val_loss']) + 1
    best_val_loss = min(history.history['val_loss'])
    print(f'\n✅ Training complete!')
    print(f'Best epoch     : {best_epoch}')
    print(f'Best val_loss  : {best_val_loss:.6f}')

    # Plot training curves
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))

    axes[0].plot(history.history['loss'],     label='Train Loss',      color='#E63946', linewidth=1.5)
    axes[0].plot(history.history['val_loss'], label='Validation Loss', color='#457B9D', linewidth=1.5)
    axes[0].axvline(best_epoch - 1, color='gray', linestyle='--',
                    label=f'Best Epoch ({best_epoch})', linewidth=1)
    axes[0].set_title('Loss Curve (MSE)', fontsize=13, fontweight='bold')
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('MSE Loss')
    axes[0].legend()

    axes[1].plot(history.history['mae'],     label='Train MAE', color='#E63946', linewidth=1.5)
    axes[1].plot(history.history['val_mae'], label='Val MAE',   color='#457B9D', linewidth=1.5)
    axes[1].set_title('MAE Curve', fontsize=13, fontweight='bold')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('MAE (scaled)')
    axes[1].legend()

os.makedirs('reports', exist_ok=True)
plt.savefig('reports/fig7_lstm_training.png',
            dpi=150,
            bbox_inches='tight')
plt.show()
```

8.EVALUATE ON TEST SET

```python
if USE_TF:
    lstm_pred_scaled = model.predict(X_test, verbose=0)
    lstm_pred = scaler.inverse_transform(lstm_pred_scaled).flatten()
    actual    = scaler.inverse_transform(y_test.reshape(-1, 1)).flatten()
    lstm_pred = np.clip(lstm_pred, 0, None)
    lstm_rmse = np.sqrt(mean_squared_error(actual, lstm_pred))
    lstm_mae  = mean_absolute_error(actual, lstm_pred)
    print(f'=== LSTM Results on TTC 30-day Test Set ===')
    print(f'RMSE : {lstm_rmse:.4f} minutes')
    print(f'MAE  : {lstm_mae:.4f} minutes')
    print(f'Mean actual delay : {actual.mean():.2f} min')
    print(f'Mean predicted    : {lstm_pred.mean():.2f} min')
```
